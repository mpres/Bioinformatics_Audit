## Part 1: Reproducing Wiggin's k-means result

Firstly, let's try to reproduce Wiggins' plot without cluster labels.

```python
sns.set_style("darkgrid")

plt.figure(figsize=(20, 14))

sns.scatterplot(data=df, x="kiwi", y="caraway", c=df["type"], edgecolor="black")

plt.xscale("log")
plt.yscale("log")

plt.show()
```

The first very obvious thing here is that the scales are logarithmic. The clusters only are apparent when using log scales. Just for good measure, let's take a look at the same plot, but with linear scales.

```python
plt.figure(figsize=(20, 14))
sns.scatterplot(data=df, x="kiwi", y="caraway", c=df["type"], edgecolor="black")
plt.show()
```

That is not such a clear definition. If we think about how K-Means works only on linearly separable data, it is obvious why it is incapable of finding the clusters here. However, it is also possible that Wiggins' algorithm had errors in it. If we implement our own version of K-Means, we should get a result that is at least similar to Wiggins'.

In order to reproduce Wiggin's result, we will create a K-Means clustering function, following the K-Means algorithm.

```python
def select_random_centroids(X: np.ndarray, n_clusters: int, random_seed: int = None):
    np.random.seed(random_seed)
    return np.copy(X[np.random.choice(len(X), size=n_clusters)])


def assign_labels(X: np.ndarray, centroids: np.ndarray):
    ret = np.full(len(X), fill_value=-1)
    for i, instance in enumerate(X):
        dist = np.linalg.norm(instance - centroids, axis=1)
        ret[i] = np.argmin(dist)
    return rankdata(ret, method="dense") - 1


def make_centroids(X: np.ndarray, cluster_labels: np.ndarray, n_clusters: int):
    centroids = np.zeros((n_clusters, X.shape[1]))
    for cluster in range(n_clusters):
        assigned_points = X[cluster_labels == cluster]
        centroids[cluster] = assigned_points.mean(axis=0)
    return centroids


def kmeans(X: np.ndarray, n_clusters: int = 5, max_iter: int = 100, random_seed: int = None):
    centroids = select_random_centroids(X, n_clusters, random_seed=random_seed)
    cluster_labels = np.full(len(X), fill_value=-1)
    for iter_step in range(max_iter):
        new_cluster_labels = assign_labels(X, centroids)
        if np.array_equal(cluster_labels, new_cluster_labels):
            break
        cluster_labels = new_cluster_labels
        centroids = make_centroids(X, cluster_labels, n_clusters=n_clusters)
    return cluster_labels, centroids
```

For a quantifiable metric of quality, we can compute the total square distance.

```python
def tot_sqdist(X: np.ndarray, centroids: np.ndarray, cluster_labels: np.ndarray):
    total = 0
    for cluster in np.unique(cluster_labels):
        cluster_points = X[cluster_labels == cluster]
        total += (np.linalg.norm(cluster_points - centroids[cluster], axis=1) ** 2).sum(axis=0)
    return total
```

Now, let's test the K-Means function on the data, and see what we find.

```python
cluster_labels, centroids = None, None
min_tot_sqdist = 1e200
for i in range(20):
    candidate_cluster_labels, candidate_centroids = kmeans(
        df[["caraway", "kiwi"]].to_numpy(), n_clusters=5,
        max_iter=1000, random_seed=i
    )
    candidate_tot_sqdist = tot_sqdist(df[["caraway", "kiwi"]].to_numpy(),
                                      candidate_centroids,
                                      candidate_cluster_labels)
    if candidate_tot_sqdist < min_tot_sqdist:
        min_tot_sqdist = candidate_tot_sqdist
        cluster_labels = candidate_cluster_labels
        centroids = candidate_centroids
```

```python
fig, ax = plt.subplots(1, 2, figsize=(20, 8))

sns.scatterplot(data=df, x="caraway", y="kiwi", c=cluster_labels, edgecolor="black", s=100, linewidth=1.2, ax=ax[0])
sns.scatterplot(x=centroids[:, 0], y=centroids[:, 1], marker="*", color="white", edgecolor="black", s=1500, linewidth=3, ax=ax[0])
ax[0].set_title("K-Means clustering (linear scale)")

sns.scatterplot(data=df, x="caraway", y="kiwi", c=cluster_labels, edgecolor="black", s=100, linewidth=1.2, ax=ax[1])
sns.scatterplot(x=centroids[:, 0], y=centroids[:, 1], marker="*", color="white", edgecolor="black", s=1500, linewidth=3, ax=ax[1])
ax[1].set_xscale("log")
ax[1].set_yscale("log")
ax[1].set_title("K-Means clustering (log scale)")
ax[1].set_xlabel("caraway (log scale)")
ax[1].set_ylabel("kiwi (log scale)")

plt.tight_layout()
plt.show()
```

This is extremely similar to Wiggins' results seen in the problem statement. Let's now try a different clustering model: the negative binomial mixture model.

## Part 2: The Negative Binomial Mixture Model

For this part, we will build a mixture model from negative binomial distributions. The dispersion $\phi$ will be a constant $0.3$ for all distributions.

### Building the mixture model

Firstly, we will define some initialization functions for our means and coefficients.

```python
def initialize_means(X: np.ndarray, n_components: int):
    means = np.zeros((n_components, X.shape[1]))
    for feature in range(X.shape[1]):
        means[:, feature] = np.random.uniform(low=X[:, feature].min(), high=X[:, feature].max(), size=n_components)
    return means


def initialize_coefficients(X: np.ndarray, n_components: int):
    coefficients = np.zeros(n_components)
    for feature in range(X.shape[1]):
        coefficients= np.random.uniform(size=n_components)
        coefficients = coefficients / sum(coefficients)
    return coefficients
```

Next, a small helper function to compute the log probability of a negative binomial distribution from a mean and a dispersion.

```python
def nbinom_log_probs(X: np.ndarray, mean: float, disp: float = 0.3):
    n = 1 / disp
    p = 1 / (1 + mean * disp)
    return nbinom.logpmf(X, n, p)
```

Now, our expectation function will compute posterior probabilities for each instance in the data for every component.

```python
def expectation(X: np.ndarray, means: np.ndarray, coefficients: np.ndarray):
    posterior_probs = np.zeros((*X.shape, len(means)))
    for feature in range(X.shape[1]):
        for component in range(len(means)):
            posterior_probs[:, feature, component] = nbinom_log_probs(X[:, feature], mean=means[component, feature])
            
    for component in range(len(coefficients)):
        posterior_probs[:, :, component] += np.log(coefficients[component] + 0.1)
    
    posterior_probs = np.exp(posterior_probs.sum(axis=1))
    posterior_probs /= posterior_probs.sum(axis=1).reshape(-1, 1)

    return posterior_probs
```

The maximization function will compute the optimal coefficients and means given posterior probabilities.

```python
def maximization(X: np.ndarray, posterior_probs: np.ndarray, means: np.ndarray, coefficients = np.ndarray):
    new_means = np.zeros_like(means)
    new_coefficients = np.zeros_like(coefficients)
    for feature in range(X.shape[1]):
        for component in range(len(coefficients)):
            posteriors_sum = sum(posterior_probs[:, component])
            new_means[component, feature] = sum(X[:, feature] * posterior_probs[:, component]) / posteriors_sum
            new_coefficients[component] = posteriors_sum / len(X)
    return new_means, new_coefficients
```

A simple function for cluster assignment will convert posterior probabilities to cluster labels.

```python
def assign_clusters(posterior_probs: np.ndarray):
    return np.argmax(posterior_probs, axis=1)
```

Lastly, the main function will implement the mixture model routine using all of our intermediate functions.

```python
def nbinom_mixture_model(X: np.ndarray, n_components: int = 5, max_iter: int = 100, random_seed: int = None):
    np.random.seed(random_seed)
    means = initialize_means(X, n_components)
    coefficients = initialize_coefficients(X, n_components)
        
    mixture_labels = np.full(len(X), fill_value=-1)
    
    for iter_step in range(max_iter):
        posterior_probs = expectation(X, means, coefficients)
        new_mixture_labels = assign_clusters(posterior_probs)
        if np.array_equal(new_mixture_labels, mixture_labels):
            break
        mixture_labels = new_mixture_labels
        means, coefficients = maximization(X, posterior_probs, means, coefficients)
        
    return mixture_labels, means, coefficients
```

### Testing the mixture model

For our mixture model, our quality metric will be the log likelihood. We will implement a function to take the log likelihood and use it to optimize the mixture model.

```python
def marginalized_feature_probas(instance: np.ndarray, n: float, ps: np.ndarray):
    return sum(nbinom.logpmf(instance[i], n, ps[i]) for i in range(len(ps)))


def log_likelihood(X: np.ndarray, coefficients: np.ndarray, means: np.ndarray, disp: float = 0.3):
    n = 1 / disp
    ps = 1 / (1 + means * disp)
    return np.sum(logsumexp(
        [[np.log(coefficients) + marginalized_feature_probas(instance, n, ps[i]) 
          for i in range(len(ps))] 
         for instance in X], axis=1))
```

With this function, we can now optimize the mixture model.

```python
mixture_labels, means = None, None
max_log_likelihood = -1e200
for i in range(20):
    candidate_mixture_labels, candidate_means, candidate_coefficients = nbinom_mixture_model(df[["caraway", "kiwi"]].to_numpy(), n_components=5, random_seed=i)
    candidate_log_likelihood = log_likelihood(df[["caraway", "kiwi"]].to_numpy(), candidate_coefficients, candidate_means)
    if candidate_log_likelihood > max_log_likelihood:
        max_log_likelihood = candidate_log_likelihood
        mixture_labels = candidate_mixture_labels
        means = candidate_means
```

```python
fig, ax = plt.subplots(1, 2, figsize=(20, 8))

sns.scatterplot(data=df, x="caraway", y="kiwi", c=mixture_labels, edgecolor="black", s=100, linewidth=1.2, ax=ax[0])
sns.scatterplot(x=means[:, 0], y=means[:, 1], marker="*", color="white", edgecolor="black", s=1500, linewidth=3, ax=ax[0])
ax[0].set_title("Mixture clustering (linear scale)")

sns.scatterplot(data=df, x="caraway", y="kiwi", c=mixture_labels, edgecolor="black", s=100, linewidth=1.2, ax=ax[1])
sns.scatterplot(x=means[:, 0], y=means[:, 1], marker="*", color="white", edgecolor="black", s=1500, linewidth=3, ax=ax[1])
ax[1].set_xscale("log")
ax[1].set_yscale("log")
ax[1].set_title("Mixture clustering (log scale)")
ax[1].set_xlabel("caraway (log scale)")
ax[1].set_ylabel("kiwi (log scale)")

plt.tight_layout()
plt.show()
```

The mixture model is able to succeed where K-Means failed. Let's compare our results with the ground truth.

```python
df["mixture_cell_type"] = mixture_labels
np.unique(df[["mixture_cell_type", "type"]].to_numpy(), return_counts=True, axis=0)
```

It seems like some pair happen much more often than others, probably due to the fact that the cluster labels themselves are arbitrary and those pairs represent the same cluster. After remapping the cluster labels to those found in the original type, we will be able to compute an accuracy metric.

```python
mapping = {
    0: 3,
    1: 1,
    2: 4,
    3: 0,
    4: 2
}
df["mixture_cell_type"] = pd.Series(mixture_labels).apply(lambda x: mapping[x])
print("Accuracy:", len(df[df["mixture_cell_type"] == df["type"]]) / len(df))
```

We can now try to find the mean expression levels of *caraway* and *kiwi* in each cell type.

```python
df.groupby("mixture_cell_type").aggregate("mean")[["caraway", "kiwi"]]
```

Lastly, let's see the proportion of each cell type in the total sample.

```python
df.groupby("mixture_cell_type").apply(lambda x: len(x) / len(df))
```

## Part 3: Fixing K-Means

We saw at the beginning that the reason K-Means fails to cluster the samples properly is because the data follow exponential scales. K-Means works best with linearly separable data. A very obvious and easy fix in this case would be to make the data linear, by running every instance through a logarithmic function.

```python
cluster_labels, centroids = None, None
min_tot_sqdist = 1e200
linear_data = np.log10(df[["caraway", "kiwi"]].to_numpy())

for i in range(20):
    candidate_cluster_labels, candidate_centroids = kmeans(linear_data, n_clusters=5, max_iter=1000, random_seed=i)
    candidate_tot_sqdist = tot_sqdist(linear_data,
                                      candidate_centroids,
                                      candidate_cluster_labels)
    if candidate_tot_sqdist < min_tot_sqdist:
        min_tot_sqdist = candidate_tot_sqdist
        cluster_labels = candidate_cluster_labels
        centroids = candidate_centroids
        
print("The best clustering has a total squared distance of", min_tot_sqdist)
```

```python
exp_centroids = 10 ** centroids

fig = plt.figure(figsize=(20, 8))
sns.scatterplot(data=df, x="caraway", y="kiwi", c=cluster_labels, edgecolor="black", s=100, linewidth=1.2)
sns.scatterplot(x=exp_centroids[:, 0], y=exp_centroids[:, 1], marker="*", color="white", edgecolor="black", s=1500, linewidth=3)
plt.xscale("log")
plt.yscale("log")
plt.title("K-Means clustering (log scale)")
plt.xlabel("caraway (log scale)")
plt.ylabel("kiwi (log scale)")

plt.show()
```

It seems that correcting for the exponential scale was enough to allow K-Means to find the correct clusters.
